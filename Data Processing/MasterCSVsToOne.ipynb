```python
import os
import glob
import pandas as pd

path = "C:/Users/Nathan/Documents/GitHub/CapstoneWifiAnalysis/Data Processing/MasterCSVs" # Set the path to the directory containing the CSV files
all_files = glob.glob(os.path.join(path, "*.csv")) # Get a list of all CSV files in the directory

df_list = [] # Create an empty list to store the individual dataframes

for file in all_files: # Loop through all CSV files
    df = pd.read_csv(file) # Read the CSV file into a dataframe
    
    df_list.append(df) # Add the dataframe to the list

combined_df = pd.concat(df_list, axis=0, ignore_index=True) # Combine all dataframes into a single dataframe
combined_df.to_csv("Master_NoTest.csv", index=False) # Save the combined dataframe to a CSV file
```

```python

#Take master csv and create a heatmap

import seaborn as sns
import matplotlib.pyplot as plt

selected_buildings = ['UREC', 'Stu', 'Atki', 'Wood', 'SVDH', 'Pros']

master = pd.read_csv('./WeekdaysAvgDiff.csv')

building_df = {}

#Seperate buildings into their own dataframes
for building in selected_buildings:
    building_df[building] = master[master['building'] == building]


#Display all dfs
for building in selected_buildings:
    display(building_df[building])



```

```python
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

#make a list of hours from 0:00:00 to 23:00:00
hours = list(range(0, 23))



# create heatmap with days as y-axis and hours as x-axis


building = 'Stu'

dfTrim = building_df[building]
dfTrim['traffic_diff_avg_zscore'] = zscore(dfTrim['traffic_diff_avg'])

#Scale traffic_diff_avg to be between 0 and 1
scaler = MinMaxScaler()
dfTrim['traffic_diff_avg_rate'] = scaler.fit_transform(dfTrim['traffic_diff_avg'].values.reshape(-1,1)) * 10





#dfTrim = dfTrim.pivot('day_name', 'time', 'traffic_diff_avg_zscore')

#Get 3 three columns of day_name, time, and traffic_diff_avg_zscore
dfTrim = dfTrim[['day_name', 'time_index', 'traffic_diff_avg_rate']]



dfTrim['day_name'] = dfTrim['day_name'].astype('category')
dfTrim['time_index'] = dfTrim['time_index'].astype('category')


pivoted = dfTrim.pivot('day_name', 'time_index', 'traffic_diff_avg_rate')

plt.title(building)

#Sort heatmap by day

pivoted = pivoted.reindex(days)

#Change label of hours to hour format
# pivoted.columns = pd.to_datetime(pivoted.columns, format='%H').time

heatmap = sns.heatmap(pivoted)
heatmap.figure.savefig("./Heatmaps/NEW" + building + ".png")


```
